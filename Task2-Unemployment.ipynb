##### 4.1 Visualizing Unemployment in India Dataset

```python
sns.pairplot(df)
```

##### 4.2 Visualizing The Labour Participation rate with the estimated employed and unemployed rates

```python
plt.figure(figsize=(10,5))
plt.title('Comparison between Estimated Unemployment Rate and Estimated Employed')
plt.xlabel('Estimated Unemployment Rate')
plt.ylabel('Estimated Employed')
sns.scatterplot(x=' Estimated Unemployment Rate (%)', y=' Estimated Employed', hue=' Estimated Labour Participation Rate (%)', data=df, s=300 , marker = '*')
```

##### From the above graph it can be predicted that Estimated Unemployment rate is high when estimated labour participation decreses whereas it is low whem estimated labour participation rate increases moderately.
##### It can also be predicted that estimated employed is maximum when estimated labour participation rate is around 20-30 % and decreases when estimated labour participation rate increases.

##### 4.3 Visualizing Estimated Unemployment Rate according to the Area

```python
fig = plt.figure(figsize = (5, 5))
sns.histplot(x=' Estimated Unemployment Rate (%)', data=df, kde=True, hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```

##### From the above graph it can be predicted that Rural areas have high counts of unemployment rate than Urban areas

##### 4.4 Correlation between features of Unemployment in India Dataset

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(5, 5))
sns.heatmap(df.corr())
plt.show()
```

##### 4.5 Visualizing Estimated Unemloyment Rate according to dates

```python
fig = plt.figure(figsize = (5, 5))
sns.lineplot(y=' Estimated Unemployment Rate (%)', x=' Date', data=df)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

##### From this graph it can be estimated that the first six months of Year 2020 has a very sharp increase in Unemployment Rate than the last six months of Year 2019 which shows no sharp increase or decrease in unemployment rate.

##### 5.2 Visualizing Estimated Labour Participation according on region

```python
fig = plt.figure(figsize = (9, 9))
plt.scatter(df1['Region.1'], df1[' Estimated Labour Participation Rate (%)'])
plt.title('Labour Participation according to Region')
plt.xlabel('Region')
plt.ylabel('Labour Participation Rate')
plt.show()
```

##### From the above graph it can be predicted that Northeast has highest labour participation rate followed by South.

##### 5.3 Visualizing Estimated Employment according to Region

```python
fig = plt.figure(figsize = (9, 9))
plt.scatter(df1['Region.1'], df1[' Estimated Employed'])
plt.title('Employment according to Region')
plt.xlabel('Region')
plt.ylabel('Employment Rate')
plt.show()
```

##### From the above grapgh it can be predicted that North has highest employment rate followed by West.

##### 5.4 Visualizing Unemployment Rate in India in the year 2020 based on different States and grouping them in different regions

```python
unemploment = df1[['Region.1', 'Region', ' Estimated Unemployment Rate (%)']]
figure = px.sunburst(unemploment, path=["Region.1", 'Region'], 
                     values=' Estimated Unemployment Rate (%)', 
                     width=700, height=700, color_continuous_scale="RdY1Gn", 
                     title="Unemployment Rate in India")
figure.show()
```

##### Plotly is an easy to use high level interface, which operates on different types of data and helps visualising data by grouping them accordingly.
##### Sunburst is used to represent each row of the dataframe as a sector
##### Unemployment rate can be obtained by hovering over the region accordingly.
